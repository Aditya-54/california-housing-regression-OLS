```python
housing_na=housing.dropna(subset=["total_bedrooms"])
housing_na.shape
```

```python
dummies=pd.get_dummies(housing_na.ocean_proximity,dtype=int)
```

```python
dummies
```

```python
housing_na_dummies = pd.concat([housing_na, dummies], axis=1)
```

```python
housing_na_dummies.head()
```

```python
housing_clean = housing_na_dummies.drop(['ocean_proximity', 'ISLAND'], axis=1)
```

```python
housing_clean.head()
```

```python
#create featues and labels
```

```python
X=housing_clean.drop(columns=["median_house_value"])
```

```python
Y=housing_clean["median_house_value"]

```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=1984)
```

```python
from sklearn.linear_model import LinearRegression
OLS=LinearRegression()
OLS.fit(X_train,Y_train)
```

```python
#Display the coefficient and intercept of OLS model
print("Intercept:", OLS.intercept_)
print("Coefficients:", OLS.coef_)
print("The R-squared value is" , OLS.score(X_train,Y_train))
```

```python
Y_pred=OLS.predict(X_test)
performance=pd.DataFrame({"Pridiction":Y_pred,"Actual_Value":Y_test})
performance['error']=performance["Actual_Value"]-performance["Pridiction"]
performance.head()
```

```python
#adding the index
performance.reset_index(drop=True, inplace=True)
performance.reset_index(inplace=True)
performance.head()
```

```python
#plotting the residuals
plt.figure(figsize=(10,6))
plt.bar('index','error',data=performance[:50],color="black",width=0.3)
plt.xlabel("Observations")
plt.ylabel("Residuals")
plt.show()
```
